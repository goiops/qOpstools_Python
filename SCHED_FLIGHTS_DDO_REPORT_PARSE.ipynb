```python
import pandas as pd
```

```python
#print('enter name of file (please include extension type)')

print('enter port abbreviation (ie.SYD, MEL, BNE, ADL)')
port = input()
filename = port+' SCHED_FLIGHTS_DDO_Report_Summary.xls'


if filename == 'MEL SCHED_FLIGHTS_DDO_Report_Summary.xls':
    data = pd.read_excel(str(filename), skiprows=5)
elif filename == 'SYD SCHED_FLIGHTS_DDO_Report_Summary.xls':
    data = pd.read_excel(str(filename), skiprows=7)
elif filename == 'BNE SCHED_FLIGHTS_DDO_Report_Summary.xls':
    data = pd.read_excel(str(filename), skiprows=8)
elif filename == 'ADL SCHED_FLIGHTS_DDO_Report_Summary.xls':
    data = pd.read_excel(str(filename), skiprows=8)
else:
    print('please check port and try again')

data
```

```python
data.dropna(subset = ["ST\n"], inplace=True)
```

```python
columns = [
 'Date',
 'ST\n',
 'Flt IN',
 'Flt OUT',
 'STA',
 'STT',
 'STD',
 'ATA',
 'ATT',
 'ATD',
 'In Gate',
 'Out Gate',
 'DDO IN\n',
 'DDO OUT\n',
 'DEPARTED < STD      \n',
 'DEPARTED < 15 min STD\n',
 'DEPARTED > 15 minSTD',
 'Arrival Occurrence\nRemark',
 'Departure Occurrence\nRemark',
]

data_cleaned = data[columns]
```

```python
column_names = [
 'Date',
 'ST',
 'Flt IN',
 'Flt OUT',
 'STA',
 'STT',
 'STD',
 'ATA',
 'ATT',
 'ATD',
 'In Gate',
 'Out Gate',
 'DDO IN',
 'DDO OUT',
 'DEPARTED < STD',
 'DEPARTED < 15 min STD',
 'DEPARTED > 15 min STD',
 'Arrival Occurrence Remark',
 'Departure Occurrence Remark'
]

data_cleaned.columns = column_names
data_cleaned['Port'] = str(port)
data_cleaned
```

```python
filename_out = port+' SCHED_FLIGHTS_DDO_output.csv'

data_cleaned.to_csv(filename_out, index = False, header=True)
```
